```python
# Choose the models you want to query
models = [
    "lfm-40b",
    "hermes-3",
    "gemini-2.0",
    "gemini-2.0-flash-thinking",
    "gemini-1.5-pro",
    "llama-3.2-3b",
    "gemini-1.5-flash",
    "glm-4",
    "blackboxai",
    "gemini-2.0-flash",
    "gpt-4",
    "llama-3.1-8b",
    "gpt-4o",
    "llama-3.1-70b",
    "evil",
    "dolphin-2.6",
    "dolphin-2.9",
    "qwen-2.5-1m",
    "blackboxai-pro",
    "qvq-72b",
    "claude-3.5-sonnet",
    "airoboros-70b",
    "sonar-pro",
    "lzlv-70b",
    "wizardlm-2-8x22b",
    "qwen-2-72b",
]

# Define your prompt
prompt = "2^10"


# To hold completed results
completed_results = []


# Ask a single model and time it
async def ask_model(client, model, prompt):
    start = time.perf_counter()
    try:
        response = await client.chat.completions.create(
            model=model, messages=[{"role": "user", "content": prompt}], web_search=False
        )
        elapsed = time.perf_counter() - start
        return model, response.choices[0].message.content, round(elapsed, 2)
    except Exception as e:
        elapsed = time.perf_counter() - start
        return model, f"Error: {str(e)}", round(elapsed, 2)


# Main runner using as_completed
async def query_models_as_completed(models, prompt):
    client = AsyncClient()

    tasks = [ask_model(client, model, prompt) for model in models]
    for coro in asyncio.as_completed(tasks):
        result = await coro
        if not result[1].startswith("Error:"):
            print(f"\n--- {result[0]} ---")
            print(f"Took {result[2]}s")
            print(result[1])
            print()
            # print(f"Model {result[0]} failed with error: {result[1]}")
            # print(len(completed_results))
            completed_results.append(result[0])
        # model_name, output, duration = result
        # print(f"\n--- {model_name} ---")
        # print(f"Took {duration}s")
        # print(output)


# Run the async flow
await query_models_as_completed(models, prompt)

# `completed_results` now holds all results in order of completion
```
